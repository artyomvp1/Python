```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Basic Graph

```python
x = [0, 1, 2, 3, 4] # координаты 
y = [0, 2, 4, 6, 8]

# Установление общего размера графика
plt.figure(figsize=(8, 5))

# Линия 1
plt.plot(x, y, color="purple", linewidth=3, marker='.', linestyle="--", markersize=15, markeredgecolor="#ababab") # стиль линии

# Линия 2
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:7], x2[:7]**2, color="red")  # Играем со срезами
plt.plot(x2[6:], x2[6:]**2, "r--") # "r--"" - красный цвет и тип линии пунктир 

# Параметры графика
plt.title("Test Title", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel("X axis", fontdict={"fontname": "Verdana"})
plt.ylabel("Y axis")
plt.legend(["1", "2"])

plt.xticks([0, 1, 2, 3, 4]) # отметки на осях
plt.yticks([0, 2, 4, 6, 8, 17])

# Сохраняем график
plt.savefig("mygraph.png") # dpi=300 - размер в пикселях

# Отобразить
plt.show()
```
